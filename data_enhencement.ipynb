```python
import os
import cv2
import json
import numpy as np
```

```python
src_img_path = 'archive/train/train'
src_labels_file = 'archive/train_annotations'

dist_img_path = 'dataset_enhenced/dataset/train/images'
dist_labels_path = 'dataset_enhenced/dataset/train/labels'

os.makedirs(dist_img_path, exist_ok=True)
os.makedirs(dist_labels_path, exist_ok=True)
```

```python
src_img_files = os.listdir(src_img_path)
src_img_files.sort()
src_img_files = list(map(lambda x: os.path.join(src_img_path, x), src_img_files))

len(src_img_files), src_img_files
```

```python
src_labels = json.loads(open(src_labels_file).read())
len(src_labels), src_labels
```

```python
def preprocess(img):
    return img
```

```python
def rotate4(img, bbox):
    
    img_height = img.shape[0]
    img_width = img.shape[1]
    
    # x_min, y_min, width, height, all in pixels
    # ==>
    # x_center, y_center, width_ratio, height_ratio
    x_min = bbox[0]
    y_min = bbox[1]
    width = bbox[2]
    height = bbox[3]
    
    x_center = (x_min + width / 2) / img_width
    y_center = (y_min + height / 2) / img_height
    width_ratio = width / img_width
    height_ratio = height / img_height
    
    yield img, [x_center, y_center, width_ratio, height_ratio]
    
    # rotate 90 degrees
    img_rotated = np.rot90(img)
    yield img_rotated, [y_center, 1 - x_center, height_ratio, width_ratio]
    
    # rotate 180 degrees
    img_rotated = np.rot90(img_rotated)
    yield img_rotated, [1 - x_center, 1 - y_center, width_ratio, height_ratio]
    
    # rotate 270 degrees
    img_rotated = np.rot90(img_rotated)
    yield img_rotated, [1 - y_center, x_center, height_ratio, width_ratio]
```

```python
counter = 0

for i in range(len(src_img_files)):
    temp_img = cv2.imread(src_img_files[i])
    temp_img = preprocess(temp_img)
    
    temp_cat = src_labels[i]['category_id'] - 1
    temp_bbox = src_labels[i]['bbox']
    for img, bbox in rotate4(temp_img, temp_bbox):
        file_name = f'image_{counter}.'
        print(file_name)
        
        cv2.imwrite(os.path.join(dist_img_path, file_name + 'jpg'), img)
        with open(os.path.join(dist_labels_path, file_name + 'txt'), 'w') as file:
            file.write(' '.join(str(value) for value in [temp_cat] + bbox))
        
        counter += 1
        
        
    img_flip = cv2.flip(temp_img, 1)
    bbox_flip = [img.shape[1] - temp_bbox[0], temp_bbox[1], temp_bbox[2], temp_bbox[3]]
    for img, label in rotate4(img_flip, bbox_flip):
        file_name = f'image_{counter}.'
        print(file_name)
        cv2.imwrite(os.path.join(dist_img_path, file_name + 'jpg'), img)
        with open(os.path.join(dist_labels_path, file_name + 'txt'), 'w') as file:
            file.write(' '.join(str(value) for value in [temp_cat] + bbox))
        
        counter += 1
        
counter
    
    
```
